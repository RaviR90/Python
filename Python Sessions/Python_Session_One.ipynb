**Naming Convetions:**

Python program contains variables, functions, classes, modules, packages, etc.

*Camel Case:*
  Each word, except the first, starts with a capital letter:
  myVariableName = "Raja"

*Pascal Case or Title Case:*
  Each word starts with a capital letter:
  MyVariableName = "Ram"

*Snake Case:*
  Each word is separated by an underscore character:
  my_variable_name = "Raja Ram"


1.   Python are case sensitive.
2.   Variables named gender and Gender are different.
3.   Class names should use the TitleCase convention. e.g. MyClass, Employee, Person.
4.   Function names should be in lowercase. Multiple words should be separated by underscores, e.g. add(num), calculate_tax(amount).
5.   Variable names in the function should be in lowercase e.g., x, num, salary.
6.   Module and package names should be in lowercase e.g., mymodule, tax_calculation. Use underscores to improve readability.
7. Constant variable names should be in uppercase e.g., RATE, TAX_RATE.

# **Python Data Types:**

Data Types are categorization of datas. Python supports the following built-in data types.

**Scalar Types:**



*   **int:** Positive or negative whole numbers. e.g. -12, 12, 456, 4654654.
*   **float:** Any real number with a floating-point representation in which a fractional component is denoted by a decimal symbol or scientific notation. e.g. 3.23, 5.4556789e2.
*   **complex:** number with a real and imaginary component. e.g x + 1y.
*   **bool:** Data with one of two built-in values True or False. e.g x = True
*   **None:** The None represents the null object in Python. e.g x = None



**Sequence Type:**

*   ***String:*** A string value is a collection of one or more characters put in single, double or triple quotes.
*   ***List:*** A list object is an ordered collection of one or more data items, not necessarily of the same type(heterogeneous), put in square brackets.
*   ***Tuple:*** A Tuple object is an ordered collection of one or more data items, not necessarily of the same type(heterogeneous), put in parentheses.


**Mapping Type:**

**Dictionary:**
A dictionary Dict() object is an ordered collection of data in a key:value pair form. A collection of such pairs is enclosed in curly brackets. For e.g: {1:"Raja", 2:"Ram", 3:"Mani", 4: "Arun"}.


**Set Types:**

*   **set:** Set is mutable, unordered collection of distinct hashable objects. The set is a Python implementation of the set in Mathematics. A set object has suitable methods to perform mathematical set operations like union, intersection, difference, etc.
*   **frozenset:** Frozenset is immutable version of set whose elements are added from other iterables.


**Immutable Types:**

Numbers, strings, and Tuples are immutable, which means their contents can't be altered after creation.

**Mutable Types: (CRUD - Create, Read, Update, Delete)**

List, Set and Dictionary object can be modified

**Python - Global Variables:**

Variables that are created outside of a function.

Global variables can be used by everywhere, both inside of functions and outside.


```python
x = "awesome"

def myfunc():
  print("Raja is " + x)

myfunc()
```

If we create a variable with the same name inside a function, this variable will be local, and can only be used inside the function.

```python
x = "awesome"

def myfunc():
  x = "fantastic"
  print("Python is " + x)

myfunc()

print("Raja is " + x)
```

To create a global variable inside a function, we can use the global keyword.

```python
def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
```
